```python
rain = ['rain', 'pouring', 'drizzle', 'drizzly', 'deluge', 'storm','Rain', 'Pouring', 'Drizzle', 'Drizzly', 'Deluge', 'Storm', 'RAIN', 'POURING', 'DRIZZLE', 'DRIZZLY', 'DELUGE', 'STORM']
fogmist = ['fog', 'mist', 'Fog', 'Mist', 'FOG', 'MIST']
snowhail = ['snow', 'hail', 'blizzard', 'frost', 'sleet' , 'Snow', 'Hail', 'Blizzard', 'Frost', 'Sleet', 'SNOW', 'HAIL', 'BLIZZARD', 'FROST', 'SLEET']
sunny = ['sunny', 'sunlight', 'sunshine', 'clear sky' , 'Sunny', 'Sunlight', 'Sunshine']
cloudy = ['gloomy', 'overcast', 'cloud', 'dark sky', 'grey sky', 'Gloomy', 'Overcast', 'Cloud', 'Dark Sky', 'Grey Sky', 'GLOOMY', 'OVERCAST', 'CLOUD', 'DARK SKY', 'GREY SKY']
thunder = ['thunder', 'lightning', 'Thunder', 'Lightning', 'THUNDER', 'LIGHTNING']
cold = ['chilly', 'cold', 'freezing', 'Chilly', 'Cold', 'Freezing', 'CHILLY', 'COLD', 'FREEZING']
warm = ['warm', 'Warm', 'WARM', 'hot', 'boiling', 'Hot', 'Boiling', 'HOT', 'BOILING']
hot = []
```

```python
def weatherf(cell_value):
    return {'rain' : sum({v for k, v in ((s, cell_value.count(s)) for s in rain) if v}), 'sunny' : sum({v for k, v in ((s, cell_value.count(s)) for s in sunny) if v}), 'cloudy' : sum({v for k, v in ((s, cell_value.count(s)) for s in cloudy) if v}), 'fog/mist' : sum({v for k, v in ((s, cell_value.count(s)) for s in fogmist) if v}), 'snow/hail' : sum({v for k, v in ((s, cell_value.count(s)) for s in snowhail) if v}), 'thunder' : sum({v for k, v in ((s, cell_value.count(s)) for s in thunder) if v})}
     
```

```python
def tempf(cell_value):
    return {'cold' : sum({v for k, v in ((s, cell_value.count(s)) for s in cold) if v}), 'warm' : sum({v for k, v in ((s, cell_value.count(s)) for s in warm) if v}), 'hot' : sum({v for k, v in ((s, cell_value.count(s)) for s in hot) if v})}
     
```

```python
def definitions(cell):
    highest = max(cell.values())
    if highest > 0:
        dictlist = [k for k,v in cell.items() if v == highest]
        return ' , '.join(str(d) for d in dictlist)
```

```python
puretweets['weathercount'] = puretweets['text'].apply(weatherf)
puretweets['tempcount'] = puretweets['text'].apply(tempf)
```

```python
puretweets['tweetweather'] = puretweets['weathercount'].apply(definitions)
puretweets['tweettemp'] = puretweets['tempcount'].apply(definitions)
```
